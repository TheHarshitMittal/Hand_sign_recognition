```python
import os
import cv2
import numpy as np
```

```python
# Input and output folders
input_folder = 'data/5'
output_folder = 'process_data/5'
```

```python
# Function to process images
def process_image(image_path):
    try:
        # Read the colored image
        colored_image = cv2.imread(image_path)

        # Convert the image to grayscale
        grayscale_image = cv2.cvtColor(colored_image, cv2.COLOR_BGR2GRAY)

        # Apply edge detection to get the borders
        edges = cv2.Canny(grayscale_image, threshold1=150, threshold2=150)

        # Save the processed image to the output folder
        output_path = os.path.join(output_folder, os.path.basename(image_path))
        cv2.imwrite(output_path, edges)
        print(f"Processed: {image_path}")
    except Exception as e:
        print(f"Error processing {image_path}: {str(e)}")
```

```python
# Process images in the input folder
for filename in os.listdir(input_folder):
    if filename.endswith('.jpg') or filename.endswith('.png'):  # Assuming images are in JPG or PNG format
        image_path = os.path.join(input_folder, filename)
        process_image(image_path)
```
